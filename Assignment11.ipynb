```python
# Q2) remove the spaces in a string

str1='hai how are you'
s=''
for i in str1:
    if i!=' ':
        s+=i
print(s)
        
```

```python
# Q3) remove digits from string
str1='hai123hello435'
s=''
for i in str1:
    if i not in '0123456789':
        s+=i
print(s)

```

```python
# Q7)count the number of vowels in a string
str1 = 'education'
count=0
for i in str1:
    if i in 'aeiou':
        count+=1
print(count)
# s = 5
```

```python
# Q10)replace spaces with underscores
str1 = 'machine learning is fun'
s=''
for i in str1:
    if i==' ':
        s+='_'
    s+=i
print(s)
```
