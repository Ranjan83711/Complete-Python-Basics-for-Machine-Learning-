```python
def addition():
    num1=eval(input('enter the no.'))
    num2=eval(input('enter the no.'))
    sub=num1-num2
    mul=num1*num2
    div=num1/num2
    print(f"subraction is : {sub}")
    print(f"multiplication is : {mul}")
    print(f"division is : {div}")
```

```python
addition()
```

```python
def area():
    r=eval(input('enter the radius'))
    pi=3.14
    area=pi*r*r
    print(f"area is : {area}")
```

```python
area()
```

```python
def area_triangle():
    h=eval(input('enter the height'))
    b=eval(input('enter the breadth'))
    area=0.5*h*b
    print(f"area of triangle is : {area}")
```

```python
area_triangle()
```

```python
def avg():
    num1=eval(input('enter the first number'))
    num2=eval(input('enter the second number'))
    num3=eval(input('enter the third number'))
    avg1=(num1+num2+num3)/3
    print(f"average of {num1} , {num2} and {num3} is : {avg1}")
```

```python
avg()
```

```python
def billing():
    bill=eval(input('enter the bill'))
    tip=eval(input('enter the tip'))
    total_amount=bill+tip
    print(f"total bill is : {total_amount}")
```

```python
billing()
```

```python
def tax_amount():
    salary=eval(input('enter the salary'))
    tax_percent=eval(input('enter the tax percent'))
    tax_amount=salary*tax_percent/100
    print(f"tax amount is : {tax_amount}")
```

```python
tax_amount()
```

```python
def rate_conversion():
    rate=80
    Dollar=eval(input('enter the amount in dollar'))
    Indian_rupees=Dollar*rate
    print(f"{Dollar} dollar is {Indian_rupees} rupees")
```
