```python
# 1 Write a Python program to find the sum of all elements in a list.
# Example:
# Input: [2, 5, 7, 1] â†’ Output: 15

lst=[2,5,7,1]
Sum=0
for i in range(len(lst)):
    Sum+=lst[i]
Sum
```

```python
# 2ï¸âƒ£ Find Maximum and Minimum

# Write a program to find the largest and smallest numbers in a list without using built-in max() and min().

lst=[2,4,7,9,12,45]
Max=float('-inf')
for i in range(len(lst)):
    if lst[i]>Max:
        Max=lst[i]
print(Max)

Min=float('inf')
for j in range(len(lst)):
    if lst[j]<Min:
        Min=lst[j]
print(Min)
    
```

```python
# 3ï¸âƒ£ Second Largest Element
# Find the second largest element in a list.
# Example:
# Input: [10, 4, 7, 9, 3] â†’ Output: 9

lst=[2,4,7,9,12,45]
largest=float('-inf')
second_largest=float('-inf')
for i in range(len(lst)):
    if lst[i]>largest:
        second_largest=largest
        largest=lst[i]
    else:
        if lst[i]!=largest:
            second_largest=lst[i]
print(second_largest)
```

# pop vs remove 
- pop takes index
- remove takes direct element name
    - taking element name advantage
    - one time element only drop
- when we want to drop second element use pop

```python
l=['apple','banana','cherry','milk','tea']
l.pop()
```

```python
l.remove('banana')
```
