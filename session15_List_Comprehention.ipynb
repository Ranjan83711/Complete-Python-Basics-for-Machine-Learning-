**List Comprehention**

- list comprehention means writing a code in a single line without using append
  
- many codes having multiple lines with conditions and loops slower the execution

- List comprehention will help you speed up the process

$Case - 1$

- focus on what is the output variable
    - which means which variable we are saving

```python
# [<output>] <for loop>] Case-1
```

```python
l=[]
for i in range(10):
    l.append(i)
l
```

```python
l=[i for i in range(10) ]
l
```

```python
Input=['hyd','bangalore','mumbai','chennai']
output=[]
for i in Input:
    output.append(i.capitalize())
output
```

```python
Input=['hyd','bangalore','mumbai','chennai']
[Input[i].capitalize() for i in range(len(Input))]
```

$Case-2$
- for loop with if condition

```python
# [<output>] <for loop>] Case-1
# [<output>] <for loop> <if condition>] Case-2
```

```python
Input=[20,21,30,31,34]
even=[]
for i in Input:
    if i%2==0:
        even.append(i)
even
```

```python
Input=[20,21,30,31,34]
even=[i for i in Input if i%2==0]
odd=[i for i in Input if i%2!=0]
print(even)
print(odd)
```

```python
Input=['hyd','bangalore','mum#bai','chen#nai']
[i for i in Input if '#' in i]
```

```python
Input=['hyd','bangalore','mum#bai','chen#nai']
[i for i in Input if not '#' in i]
```

$Case - 3$ : for-if-else

```python
# [<output>] <for loop>] Case-1
# [<output>] <for loop> <if condition>] Case-2
# [<if_output> <if condition> else <else_output> <for loop>] Case-3
```

```python
Input=[20,21,30,31,34]
[f"even_{i}" if i%2==0 else f"odd_{i}" for i in Input]
```

$Case-4$ : if-elif-else

```python
# [<output>] <for loop>] Case-1
# [<output>] <for loop> <if condition>] Case-2
# [<if_output> <if condition> else <else_output> <for loop>] Case-3
# [if_output] <if condition> else <elif output> <elif condition> else <else output> <for loop>] Case-4

# Rule-1: There is no elif
        # else-if
# Rule-2 : if---else {always side by side}
# Rule-3 : else ke baad pakka output aata
```

```python
Input=[50,70,90]
['A' if i>=90 else 'B' if i>=60 else 'C' for i in Input]
```

```python
Input=[50,70,90]
['A' if i>=90\
 else 'B' if i>=60\
 else 'C'\
 for i in Input]
```

```python
# 1. Square Even Numbers Only
# Given a list of integers, create a new list containing the squares of only the even numbers.

nums = [2, 5, 8, 11, 14, 17, 20]
l=[i*i for i in nums if i%2==0]
l
```

```python
# 2. Filter Strings with More Than 3 Vowels
# From a list of words, return only those that have more than 3 vowels.

words = ["education", "sky", "queue", "beautiful", "crypt"]
vowels='aeiouAEIOU'
# [i for i in words if (len(i) in vowels)>3]
# result = [i for i in words 
#           if len([j for j in i if j in vowels]) > 3]
# result

s=[]
for i in words:
    count=0
    for j in i:
        if j in vowels:
            count+=1
    if count>3:
        s.append(i)
print(s)
           
```

```python
# 3. Flatten a 2D List
# Flatten a 2D list into a 1D list.

lst = [[1, 2], [3, 4], [5, 6]]
# [x for i in lst for x in (i if type(i)==list else [i])]
a=[lst[i][j] for i in range(len(lst)) for j in range(len(lst[i]))]
# a = [lst[i][j] for i in range(len(lst)) for j in range(len(lst[i]))]
a
```

```python
# 4. Get First Letters of Words
# Extract the first letter from each word in a sentence.

sentence = "Python list comprehension is powerful"
[i[0] for i in sentence.split()]
```

```python
# 5. Filter Palindromes
# Return only palindromic strings from a given list.

words = ["madam", "python", "level", "noon", "data"]
[i for i in words if i==i[::-1]]
# a='data'
# a[::-1]
```

```python
# 6. Replace Negative Numbers with 0
# Replace all negative numbers in a list with 0.

nums = [-2, 3, -5, 7, -9, 10]
[0 if i<0 else i for i in nums]
```

```python
# 7. Generate (number, square, cube) Tuples
# Generate a list of tuples containing a number, its square, and its cube for numbers 1â€“10.
# tup=[i for i in range(0,11)]
# square=[i*i for i in tup]
# cube=[i*i*i for i in tup]
# print(tup,square,cube)
tup=[(i,i*i,i*i*i) for i in range(0,11)]
print(tup)
```

```python
# 8. Words Containing a Specific Letter
# Extract all words containing the letter 'a' from a sentence.

sentence = "Practice makes a man perfect"
[i for i in sentence.split() if 'a' in i]
```
