```python
l1=[10,20,30,40]
l2=[1,2,3]
l=[]
if len(l1)>len(l2):
    for i in range(len(l1)):
        if i<len(l2):
            l.append(l1[i]+l2[i])
        else:
            l.append(l1[i])  
else:
    for i in range(len(l2)):
        if i<len(l1):
            l.append(l1[i]+l2[i])
        else:
            l.append(l2[i])  
    
l
```

```python
from itertools import zip_longest
```

```python
l1=[10,20,30,40]
l2=[1,2,3]
list(zip(l1,l2))
```

```python
l1=[10,20,30,40]
l2=[1,2,3]
list(zip_longest(l1,l2,fillvalue=0))
```

```python
l1=[10,20,30,40]
l2=[1,2,3]
l=[]
new=list(zip_longest(l1,l2,fillvalue=0))
for i,j in new:
    l.append(i+j)
l
```

**List to dictionary conversion**

```python
name=['ramesh','suresh','mahesh']
age=[20,30,40]
d={}
for i,j in zip(name,age):
    d[i]=j
d
```

```python
numbers=[20,21,22]
d={}
for i in numbers:
    if i%2==0:
        d[i]='even'
    else:
        d[i]='odd'
d    
```
