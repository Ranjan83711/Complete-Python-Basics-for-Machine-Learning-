```python
# Q1) [2,3,30,40,50,6,7]
#    find the maximum number in a list with out using max function
lst=[2,3,30,40,50,6,7]
Max=float('-inf')
for i in range(1,len(lst)):
    Max=max(Max,lst[i])
print(Max)
```

```python
# Q2) [2,3,30,40,50,6,7]
#    find the second maximum number in a list with out using max function
lst=[2,3,30,40,50,6,7]
Max=float('-inf')
sec_Max=float('-inf')
for i in range(0,len(lst)):
    if lst[i]>Max:
        sec_Max=Max
        Max=lst[i]
    elif lst[i]>sec_Max and lst[i]!=Max:
        sec_Max=lst[i]
print(sec_Max)
```

```python
# Q3) [2,3,30,40,50,6,7]
#     sort the values with out sort method

lst = [2, 3, 30, 40, 50, 6, 7]
for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] > lst[j]:
            lst[i], lst[j] = lst[j], lst[i]  
print(lst)
    
```

```python
# Q4) [2,3,30,40,50,6,7]
#     reverse the values with out reverse method

lst = [2, 3, 30, 40, 50, 6, 7]
n=len(lst)
first=0
last=n-1
while first<last:
    lst[first],lst[last]=lst[last],lst[first]
    first+=1
    last-=1
lst
```

```python
# Q8) str1='hello how are you, hello im good'
#     take any sentence above find the most reptaed word
from collections import Counter

str1 = 'hello how are you, hello im good'

# Step 1: clean text
str1 = str1.replace(',', '').lower()

# Step 2: split into words
words = str1.split()

# Step 3: use Counter to count words
word_count = Counter(words)

# Step 4: find the most repeated word
max_word, max_count = word_count.most_common(1)[0]

print("Most repeated word:", max_word)
print("Count:", max_count)
```

```python
str1 = 'hello how are you, hello im good'

# remove punctuation and make lowercase
str1 = str1.replace(',', '').lower()

# split into words
words = str1.split()

most_word = ''
max_count = 0

for word in words:
    count = 0
    for w in words:
        if word == w:   # simple 'if' comparison
            count += 1
    if count > max_count:
        max_count = count
        most_word = word

print("Most repeated word:", most_word)
print("Count:", max_count)
```

```python
# Q13) [202,301,405,607,590,320,140,50]   T
#      [102,201,305,407,390,220,140,50]   T
#      [102,201,305,407,390,220,440,50]  False
#    you need to create a function
#    the function should verify the above pattern with any list

# lst=[202,301,405,607,590,320,140,50]  
lst=[102,201,305,407,390,220,440,50]
def solution(lst):
    maxx=float('-inf')
    n=len(lst)
    for i in range(n):
        if maxx<lst[i]:
            maxx=lst[i]
            max_index=i
    for i in range(1,max_index+1):
       if lst[i]<lst[i-1]:
           return False
    for i in range(max_index+1,n):
        if lst[i]>=lst[i-1]:
            return False
    return True
solution(lst)
```
