# **Type Cast**

- Convert one data type to anothet data type 

```python
a=100
b=10
c=a+b
c
```

```python
a='100'
b='10'
c=a+b
c
```

- in above both strings are concatenated
- if we wants a correct answer we need to convert str to int
- this is one example of typecast
