```python
# Q1: Sum of Three Numbers
# Write a function `sum_three(a, b, c)` that returns the sum of the three numbers.
# Example:
# sum_three(3, 5, 7) ➞ 15

def sum_three(a,b,c):
    Sum=a+b+c
    print(f"The sum of {a},{b} and {c} is : {Sum}")

sum_three(3, 5, 7)
```

```python
# Q2: Area of a Circle
# Write a function `area_circle(radius)` that returns the area of a circle. Use π = 3.14159
# Example:
# area_circle(7) ➞ 153.9381
import math
def area_circle(r):
    area=round(math.pi*r**2,2)
    print(f"arae of the circle with radius {r} is : {area}")
area_circle(7)
```

```python
# Q3: Check Even or Odd
# Write a function `check_even_odd(num)` that prints whether a number is Even or Odd.
# Example:
# check_even_odd(5) ➞ &quot;Odd&quot;
# check_even_odd(8) ➞ &quot;Even&quot;

def check_even_odd(n):
    if n%2==0:
        print(f"given number : {n} => even num")
    else:
        print(f"given number : {n} => odd num")

check_even_odd(5)
check_even_odd(8)
```

```python
# Q4: Calculate BMI
# Write a function `calculate_bmi(weight, height)` that returns BMI value using the formula:
# BMI = weight / (height ** 2)
# Example:
# calculate_bmi(70, 1.75) ➞ 22.86

def calculate_bmi(weight,height):
    BMI=round(weight/(height**2),2)
    print(f"BMI for the given weight:{weight} and height:{height} is {BMI}")
calculate_bmi(80, 6)
```

```python
# Q5: Greet with Name and Age
# Write a function `greet(name, age)` that prints:
# Hello &lt;name&gt;, you are &lt;age&gt; years old!

# Example:
# greet(&quot;Omkar&quot;, 30) ➞ Hello Omkar, you are 30 years old!

def greet(name,age):
    print(f"Hello {name}, you are {age} yeras old!")
greet('Ranjan',25)
```

```python
# Q6: Find Maximum of Two Numbers
# Write a function `find_max(a, b)` that returns the greater of the two numbers.
# Example:
# find_max(4, 9) ➞ 9

def max_num(a,b):
    temp=max(a,b)
    print(f"the maximum num between {a} and {b} is : {temp}")
max_num(6,4)
```

```python
# Q7: Power Function
# Write a function `power(base, exponent)` that returns base raised to the power exponent.
# Example:
# power(2, 3) ➞ 8

def power(a,b):
    Power=pow(a,b)
    print(f"with base : {a} and power : {b}, the number becomes : {Power}")
power(2, 3)
```

```python
# Q8: Calculate Simple Interest
# Write a function `simple_interest(principal, rate, time)` that returns the simple interest.
# Formula: (P × R × T) / 100
# Example:
# simple_interest(1000, 5, 2) ➞ 100.0

def simple_interest(p,r,t):
    SI=round((p*r*t)/100,2)
    print(f"the SIMPLE INTEREST with principal:{p},rate:{r} and time:{t} is : {SI}")
simple_interest(5000,4,1)
```

```python
# Q9: Check Leap Year
# Write a function `is_leap_year(year)` that returns True if the year is a leap year, otherwise
# False.
# Example:
# is_leap_year(2020) ➞ True
# is_leap_year(2023) ➞ False

def is_leap_year(year):
    if ((year%4==0 and year%100!=0) or year%400==0):
        print(f"Given year:{year} is a leap year")
    else:
        print(f"Given year:{year} is not a leap year")

is_leap_year(2026)
is_leap_year(2025)
is_leap_year(2024)
```
