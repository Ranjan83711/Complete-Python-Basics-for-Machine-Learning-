# finally block


- **try block** is used to run original code or correct code
- **except block** is used to catch the errors mainly
- in cloud whenever we establish a connection the money deduction starts
- so even for correct code or wrong code the money deduction starts when connection establish
- so it is very important to close the connection
- in such condition we use **finally** block
- if **try block** executes entirely after **finally block** will also be execute
- if **except block** executes entirely, **finally block** also executes
- in both conditions **finally block** will executes

```python
try:
    num1=eval(input("enter the number1 : "))
    num2=eval(input("enter the number2 : "))
    avg=(num1+num2)/2
    print(f"the average of {num1} and {num2} is {avg}")

except Exception as e:
    print(e)
    print("ha ha")

finally:
    print('bye')
```
