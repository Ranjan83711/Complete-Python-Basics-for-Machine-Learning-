```python
import random
i=1
Even=0
Odd=0
while(i<6):
    num=random.randint(1,100)
    print(num)
    if num%2==0:
        Even=Even+num
        print("even")
    else:
        Odd+=num
        print("odd")
    i=i+1
print(f"sum of even is : {Even}")
print(f"sum of odd is : {Odd}")
    
    
```
