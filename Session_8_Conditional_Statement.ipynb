**if-elif-else**

- When ever we have one condition two outputs then we choose if-else
    - if need condition
    - else does not need condition

- when we have two conditions 3 outputs
    - if need conditions 
    - elif need conditions 
    - else does not need conditions 

```python
num=int(input('enter the number'))
if num==1:
    print('one')

elif num==2:
    print('one')

elif num==3:
    print('three')

else:
    print('invalid number')
```

```python
age=eval(input('enter the age'))
gender=input('enter the gender')
if gender=='male':
    if age>=100:
        print('Unlucky man')

    elif age>=60:
        print('Senior citizen')

    elif age>=35:
        print('Middle age man')

    elif age>=20 :
        print('Young man')

    elif age>=10:
        print('boy')

    else:
        print('kid')

elif gender=='female':
    if age>=100:
        print('Unlucky woman')

    elif age>=60:
        print('Senior citizen woman')

    elif age>=35:
        print('Middle age woman')

    elif age>=20 :
        print('Young girl')

    elif age>=10:
        print('girl')

    else:
        print('kid')

else:
    print('enter a valid gender')
```
