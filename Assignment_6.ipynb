```python
# 1.Write a program to check if a number entered by the user is even or odd.

num=eval(input("enter the number"))
if num%2==0:
    print(f"{num} is even")

else : print(f"{num} is odd")
```

```python
# 2.Write a program to check whether a number is positive, negative, or zero.

num=eval(input("enter the number"))
if num==0:
    print(f"{num} is zero")

elif num>0:
    print(f"{num} is positive")

else : print(f"{num} is negative")
    
```

```python
# 3.Write a program that takes two numbers and prints the larger one.

num1=eval(input("enter the first number"))
num2=eval(input("enter the second number"))
if num1>num2:
    print(f"largest number : {num1}")

else: print(f"largest number : {num2}")
```

```python
# 4.Ask the user to enter their age and check if they are eligible to vote (age >= 18).

age=eval(input("enter the age"))
if age>=18:
    print("Eligible for vote")

else: print(" Not Eligible for vote")

```

```python
# 5.Check if a number is divisible by both 5 and 11.

num=eval(input("enter the number"))
if num % 5 == 0 and num % 11 == 0:
    print(f"{num} : divisible by both 5 and 11")

else : print(f"{num} : not divisible by both 5 and 11")
```
