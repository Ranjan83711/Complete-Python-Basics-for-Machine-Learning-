```python
# wap ask the user take height and breadth find the area of traingle

h=eval(input('enter the height'))
b=eval(input('enter the breadth'))
area=0.5*h*b
print(f"area of triangle is : {area}")
```

```python
# wap ask the user take two number print subtraction multiplication and division

num1=eval(input('enter the no.'))
num2=eval(input('enter the no.'))
sub=num1-num2
mul=num1*num2
div=num1/num2
print(f"subraction is : {sub}")
print(f"multiplication is : {mul}")
print(f"division is : {div}")
```

```python
# wap ask the user take 3 numbers find the average

a=eval(input('enter the first no.'))
b=eval(input('enter the second no.'))
c=eval(input('enter the third no.'))
avg=(a+b+c)/3
print(f"average of {a}, {b} and {c} is : {round(avg,2)}")
```

```python
# wap ask the user take the bill amount and tip amount calculate total bill

bill=eval(input('enter the bill'))
tip=eval(input('enter the tip'))
total_amount=bill+tip
print(f"total bill is : {total_amount}")
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount

salary=eval(input('enter the salary'))
tax_percent=eval(input('enter the tax percent'))
tax_amount=salary*tax_percent/100
print(f"tax amount is : {tax_amount}")
```
