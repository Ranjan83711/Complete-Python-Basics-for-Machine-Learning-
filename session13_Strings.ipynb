```python
#Write a program to count how many times each character appears in a given string0

def count_letters(s):
    checked = ''  # to keep track of letters we've already counted
    for ch in s:
        if ch not in checked:
            count = 0
            for c in s:
                if c == ch:
                    count += 1
            print(f"{ch} : {count}")
            checked += ch

string = "hello world"
count_letters(string)
```

```python
# Write a program to capitalize the first letter of each word in a string without using the built-in title() function.

text='india vs aus odi series start from 19th oct'
result = ""               # to store the final string
capitalize_next = True    # flag to capitalize the first character

for ch in text:
    if capitalize_next and 'a' <= ch <= 'z':   # check if we should capitalize
        result += chr(ord(ch) - 32)            # convert lowercase to uppercase
        capitalize_next = False
    else:
        result += ch                           # add character as is

    # if we find a space, next letter should be capitalized
    if ch == ' ':
        capitalize_next = True

print(result)
```
