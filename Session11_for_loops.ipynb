# LOOPS
- repeating multiple times
- any loop we need 3 values
    - start initialization
    - Direction  = increment by 1 or decrement by 1
        - positive direction : increment
        - negative direction : decrement
    - stop : condition to exit 

**TYPES**
- for 
- while

**range(stop)**

- direction= positive

$CASE-2$ : **range(start,stop)**

- if we want to start a loop from a specific value we can mention start point also
- start=start
- direction=positive
    - step by default increment by 1
- last=stop-1

```python
range(5,10)
```

```python
for i in range(5,20):
    print(i,end=' ')
```

```python
# wap ask the user get 5 random numbers
# print even or odd
import random
def even_odd():
    n=random.randint(1,100)
    if n%2==0:
        print(f"{n} is even")
    else:
        print(f'{n} is odd')
for i in range(5):
    even_odd()
    
```

# SUMMATION WRAPPER
A Summation Wrapper is a loop that continuously adds values to compute the total sum of a series of numbers.

```python
# wap ask the user sum of first 10 natural numbers
Sum=0
for i in range(1,11):
    Sum=Sum+i
print(Sum)
```

```python
s=0
for i in range(5):
    num=random.randint(1,100)
    print(num)
    if num%2==0:
        s=s+num
        print("even")
print(s)
    
```

```python
even_s=0
odd_s=0
for i in range(5):
    num=random.randint(1,100)
    print(num)
    if num%2==0:
        even_s+=num
        print("even")
    else:
        odd_s+=num
        print("odd")
print(f"sum is {even_s}")
print(f"sum is {odd_s}")
    
```

```python
n=75
Add=0
count=0
for i in range(1,n+1):
    if n%i==0:
        count=count+1
        print(i)
        Add=Add+i
print(Add)
print(f"count is {count}")
        
```

# Counter Wrapper
A Counter Wrapper is a loop structure used to count something repeatedly —
for example, counting how many numbers satisfy a condition.

```python
n=75
Add=0
count=0
for i in range(1,n+1):
    if n%i==0:
        count=count+1
        print(i)
        Add=Add+i
print(Add)
print(f"count is {count}")
```
