```python
# Important nuances

# Unwinding still happens even if you donâ€™t use the return value. Python will return (explicitly or implicitly None) and pop the stack.
# So unwinding is always part of the execution model.

# If a function does work only during winding down, the â€œvisibleâ€ action is happening while descending; unwinding is silent but still occurs.

# Tail-call optimization (TCO): In some languages with TCO, a tail-recursive call can be optimized into a loop so the stack doesnâ€™t grow; 
# that changes the implementation (stack frames may not be created) but conceptually you still have the idea of repeating the step until the base case 
# â€” the separate â€œphasesâ€ are optimized away. Python does not do TCO, so winding and unwinding are real and observable in Python.
```

```python
def power(a,b):
    if a==0:
        return 1
    POW=a*power(a,b-1)
    return POW
power(1,3)
```

```python
def ADD(n):
    if n==0:
        return 1
    SUM=n+ADD(n-1)
    return SUM
ADD(5)
```

```python
def cal_sum2(n):
    if n==0:
        return 0
    SUMM=n%10 + cal_sum2(n//10)
    return SUMM
cal_sum2(1234)
```

```python
# Write a recursive function to calculate a^b (power function).

def Power(a,b):
    if b==0:
        return 1
    Product=a*Power(a,b-1)
    return Product
Power(3,3)
```

```python
# Write a recursive function to find the greatest common divisor (GCD) of two numbers.

def GCD(a,b):
    if a==0 or b==0:
        return 1
    div=b%a
    return div

GCD(48,56)
        
    
```
