**Lambda funcion**

- it is a function we can represent in a single line
- it behaves exactly how our original function does
- lambda arguments : expression 

**case-1 : function with one arguments**

```python
def add(n):
    return(n+10)
add(5)
```

```python
# syntax : <function name> = lambda <arg> : <return value>
# function name=add
# arg : n
# return : n+10
```

```python
add1=lambda n:n+10
```

```python
add1(5)
```

```python
def cube(a):
    return(a*a*a)
cube(2)
```

```python
cube=lambda a:a*a*a
cube(3)
```

**case-2 : more than one argument**

```python
summ=lambda n1,n2:n1+n2
summ(2,3)
```

```python
avg=lambda n1,n2,n3:(n1+n2+n3)//3
avg(5,10,15)
```

```python
avg1=lambda a,b,c:round((a+b+c)//3)
avg1(5,20,30)
```

```python
avg1=lambda a,b,c=10:round((a+b+c)//3)
avg1(5,20)
```

**Case-3 : if-else statement**

```python
<fun name>= lambda <arg> : <expression>
<fun name>= lambda <arg1> , <arg2> : <expression>
<fun name>= lambda <arg1> : <if_out> <if cond> else<else_out>
```

```python
even_odd = lambda i : 'even' if i%2==0 else 'odd'
even_odd(10)
```

```python
even_odd = lambda i : f'even_{i}' if i%2==0 else f'odd_{i}'
even_odd(15)
```

```python
greater = lambda a,b : a if a>b else b
greater(50,10)
```

```python
grade = lambda n : 'A' if n>=75 else 'B' if n>=60 else 'C'
grade(50)
```

# reduce

- reduce is a method will do all math operation with lambda func
- reduce will available package called as **functools**
- import functools and apply dir

```python
summ=0
[summ:=summ+i for i in range(1,11)][-1]
```

```python
l=[1,2,3,4]
map(lambda i,summ :summ+i,l)
```

map - in built
filter - in built
reduce - method 

```python
lambda summ,i: summ+i,l
```

```python
from functools import reduce
l=[1,2,3,4]
reduce(lambda summ,i: summ+i,l)
```

```python
l=[10,20,30,40]
reduce(lambda s,i: s+i,l)  # by default it takes the value of s=0
```

```python
l=[10,20,30,40]
reduce(lambda s,i: s+i,l,100)  # we can also assign the value from where we want to start
```
