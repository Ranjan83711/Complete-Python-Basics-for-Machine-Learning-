- in recursive fun a stop condition mandatory
- if we dont provide the conditions , it will infinite
- if same logic repeting multiple times then think about recursion 

```python
def num(n):
    if n==0:
        return 0
    num(n-1)
    print(n)
num(10)
```

```python
def Sum(n):
    if n==0:
        return 0
    Add=n + Sum(n-1)
    print(Add)
    return Add

Sum(5)
```

```python
def factorial(n):
    if n==0:
        return 1
    fact=n * factorial(n-1)
    return fact

factorial(5)
```

```python
def rec(x,n):            #parametrized recurssion
    if n==0:
        return 0
    print(x)
    rec(x,n-1)
rec(10,4)
```

```python
def fibonacci_series(n):
    if n==0 or n==1:
        return n
    Sum=fibonacci_series(n-1)+fibonacci_series(n-2)
    return Sum
fibonacci_series(5)
```
