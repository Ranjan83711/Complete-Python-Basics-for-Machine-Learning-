# WHILE LOOP

- $WHILE LOOPS$ are used for unlimited executions

- for examples we don't know when to stop the loop

- we don't know the condititon to stop

- we have requirements the loop will run unlimited till we get True value then go for the $while loop$

- if we know how many times the loop can run - go with **for loop**

- $wile loops$ works for limited executions like for loop, also unlimited executions 

- initial point
- condition to stop
- increment or decrement

```python
# start
# while<condition>:
#     <code line1>
#     <code line2>
#     increment or decrement
```

```python
for i in range(1,10):
    print(i)
```

```python
i=1
while(i<10):
    print(i,end=' ')
    i=i+1



# i=1 i>2 False No entry
# i=1 i<2 False entry
```

```python
for i in range(10,0,-1):
    print(i,end=' ')
```

```python
for i in range(-1,-11,-1):
    print(i,end=' ')
```

```python
i=10
while(i>0):
    print(i,end=' ')
    i-=1
```

```python
i=-1
while(i>-11):
    print(i,end=' ')
    i=i-1
```

```python
i=1
while True:
    print(i,end=' ')
    i=i+1
    if i==10:
        break
```

```python
i=-1
while True:
    print(i,end=' ')
    i=i-1
    if i==-10:
        break
```

```python
i=10
while True:
    print(i,end=' ')
    i-=1
    if i==1:
        break
```

```python
import random
i=1
while(i<=3):
    random_number=random.randint(1,10)
    user_num=eval(input("enter the num"))
    print(random_number)
    if random_number==user_num:
        print("Won")
    else:
        print("lost")

    i=i+1
    
```
